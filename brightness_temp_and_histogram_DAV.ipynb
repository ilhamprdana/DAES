# Import Modules

```python
import xarray as xr  
import numpy as np
from matplotlib import pyplot as plt
import scipy
import pandas as pd
import matplotlib.dates as mdates
plt.rcParams['font.sans-serif'] = 'Helvetica'
```

# Define Functions

```python
def cal_grad(array): 
    """
    This function will calculate the gradient angle of 2D image array (brightness temperature)
    Steps:
        1. Calculate the gradient in X and Y
        2. Calculate the magnitude
        3. Calculate the angle of the vector using arc tan and convert it into 0 - 360 degree
    
    """ 
    # This function will calculate the gradient in X and Y direction, magnitude, and degree
    Gx = scipy.ndimage.sobel(array, axis=1)     # Find the gradient in X direction
    Gy = -1*scipy.ndimage.sobel(array, axis=0)  # Find the gradient in Y direction
       
    # Calculate the magnitude
    G_mag       = np.sqrt(Gx**2+Gy**2)          # Find the magnitude
    G_angle     = np.arctan2(Gy,Gx)*180/np.pi   # Find the angle of vector (can be negative)
            
    G_angle_360 = np.where(G_angle>=0, G_angle, G_angle+360)    # Convert negative angle to positive
    
    return G_mag, G_angle_360

def cal_dev_angle(array, center=None):
    """
    Calculate DAV
    Input    : 
        1. Array 2D Vector Gradient Angle in every shape array, size of the interest area in pixels
        2. center = center of the reference point, if None so the center will be in the center of array
        3. radius = 
    
    Remember : column means Y, row means X; reference point is the center of the array
    Output   : Deviation Angle of the interest area related to reference point
    
    """ 
    if center is None: # use the middle of the image
        center = (int(np.shape(array)[0]/2), int(np.shape(array)[1]/2))
   
    Y, X = np.ogrid[:np.shape(array)[0], :np.shape(array)[1]]
    
    rad_line = (180/np.pi)*np.arctan2((center[0]-Y),(X-center[1]))
    rad_line = np.where(rad_line>=0, rad_line, rad_line+360)
    
    # Find the difference between vector input angle and the radial line
    dif_angle = array - rad_line
    
    dif_angle = np.where((dif_angle >= 270) & (dif_angle <=  360), dif_angle - 360, dif_angle)
    dif_angle = np.where((dif_angle >   90) & (dif_angle <   270), dif_angle - 180, dif_angle)
    dif_angle = np.where((dif_angle > -270) & (dif_angle <   -90), dif_angle + 180, dif_angle)
    dif_angle = np.where((dif_angle >=-360) & (dif_angle <= -270), dif_angle + 360, dif_angle)
    
    return dif_angle

def mask_BT(array, radius=None):
    """
    Make a mask for the brightness temperature data so make it become circular
    
    """
    # Create Masked BT 
    mask = circular_mask(np.shape(array)[0], np.shape(array)[1], radius=radius)
    masked_array = array.copy()
    masked_array[~mask] = np.nan
    
    return masked_array

def circular_mask(h, w, center=None, radius=None):
    """
    Create circular mask
    
    """  
    if center is None: # use the middle of the image
        center = (int(h/2), int(w/2))
    if radius is None: # use the smallest distance between the center and image walls
        radius = min(center[0], center[1], w-center[0], h-center[1])
        
    X,Y = np.ogrid[:h, :w]
    dist_from_center = np.sqrt((X - center[0])**2 + (Y-center[1])**2)

    mask = dist_from_center <= radius
    return mask

def cal_dav(array, center=None, radius=30): 
    """
    Calculate Deviation Angle variance based on the specified pixels size
    Input  : Array of deviation angle
    Output : DAV Value
    
    """
    # Cropped the interest area
    mask = circular_mask(np.shape(array)[0],np.shape(array)[1], center=center, radius=radius)
    dev_mask = array.copy()
    dev_mask[~mask] = np.nan    

    del mask
    
    return dev_mask, np.nanvar(dev_mask)
```

# Read Data

The microwave data used in this study is obtained from http://tropic.ssec.wisc.edu/real-time/mimtc/. The data is a morphed images of passive microwave satellite from several sensors. See Wimmers and Velden (2007) for more details about the datasets. [https://doi.org/10.1175/BAMS-88-8-1187]

```python
data1 = xr.open_dataset('./data/20231009T000000.nc')
lon1  = data1.longitude[2:203]
lat1  = data1.latitude
bt1   = data1.mimic_tc_89GHz_bt[2:203,:]

data2 = xr.open_dataset('./data/20231012T000000.nc')
lon2  = data2.longitude[6:207]
lat2  = data2.latitude
bt2   = data2.mimic_tc_89GHz_bt[6:207,:]

data3 = xr.open_dataset('./data/20231013T120000.nc')
lon3  = data3.longitude[14:215]
lat3  = data3.latitude
bt3   = data3.mimic_tc_89GHz_bt[14:215,:]
```

Read the Tropical Cyclone Data and DAV value that has been calculated in particular radius stored in excel files, which is 150 km in this case.

```python
data_tc = pd.read_excel('./data/Bolaven_DAV_150_KM.xlsx')
data_tc = data_tc.replace(-9999, np.nan)
```

# Calculation

Calculate the deviation angles of the brightness temperature within the radius of 150 km. Deviation angle variance (DAV) value here is calculated based on the method explained by Ritchie et al. (2012) [https://doi.org/10.1175/WAF-D-11-00156.1] 

```python
radius_eval = 30 #radius in pixels, 1 pixels = 0.05° x 0.05° which are about 5 km.

grad_magnitude1, grad_angle1 = cal_grad(bt1)
dev_angle1 = cal_dev_angle(grad_angle1)
dev_angle_r30_1 = mask_BT(dev_angle1, radius=radius_eval)
dav_value_r30_1 = cal_dav(dev_angle1, radius=radius_eval)[1]

grad_magnitude2, grad_angle2 = cal_grad(bt2)
dev_angle2 = cal_dev_angle(grad_angle2)
dev_angle_r30_2 = mask_BT(dev_angle2, radius=radius_eval)
dav_value_r30_2 = cal_dav(dev_angle2, radius=radius_eval)[1]

grad_magnitude3, grad_angle3 = cal_grad(bt3)
dev_angle3 = cal_dev_angle(grad_angle3)
dev_angle_r30_3 = mask_BT(dev_angle3, radius=radius_eval)
dav_value_r30_3 = cal_dav(dev_angle3, radius=radius_eval)[1]
```

# Plot

Plot the brightness temperature of the tropical cyclone Bolaven cases in three particular times that represents different TC stages along with the histogram of the deviation angles (with DAV value) that corresponds with the brightness temperature. In addition, we will also plot the time series of the 24hour backward averaged DAV. See Hu et al. (2020) regarding the averaging. [https://doi.org/10.1175/WAF-D-19-0102.1]

```python
draft = True #True for drafting, and False for printing

fs = 24
fs_i = 28

plt.rcParams.update({'font.size': fs}) 
fig = plt.figure(figsize=(16, 12))


ax1 = fig.add_axes([0.1, 0.725, 0.22, 0.293])
ax1.pcolor(lon1, lat1, bt1.T, cmap='jet_r', vmin=160, vmax=280)
ax1.set_xlabel('longitude ($\\degree$)')
ax1.set_ylabel('latitude ($\\degree$)')
ax1.set_title('a', loc='left', fontweight='bold', fontsize=fs_i)
ax1.set_title('55 knots', loc='right', fontweight='bold', fontsize=fs)

ax2 = fig.add_axes([0.4, 0.725, 0.22, 0.293])
ax2.pcolor(lon2, lat2, bt2.T, cmap='jet_r', vmin=160, vmax=280)
ax2.set_xlabel('longitude ($\\degree$)')
ax2.set_title('b', loc='left', fontweight='bold', fontsize=fs_i)
ax2.set_title('115 knots', loc='right', fontweight='bold', fontsize=fs)

ax3 = fig.add_axes([0.7, 0.725, 0.265, 0.293])
c3 = ax3.pcolor(lon3, lat3, bt3.T, cmap='jet_r', vmin=160, vmax=280)
ax3.set_xlabel('longitude ($\\degree$)')
ax3.set_title('c', loc='left', fontweight='bold', fontsize=fs_i)
ax3.set_title('90 knots', loc='right', fontweight='bold', fontsize=fs)

fig.colorbar(c3, ax=ax3, location='right', label='Brightness temperature (K)', extend='both', pad=0.05)

ax4 = fig.add_axes([0.1, 0.45, 0.22, 0.15])
ax4.hist(dev_angle_r30_1.flatten(), bins=np.arange(-90, 92.5, 2.5), 
         edgecolor='black', color='black', align='mid', rwidth=1.0)
ax4.set_xlim(-90,90)
ax4.set_xlabel('deviation angle ($\\degree$)')
ax4.set_ylabel('frequency')
ax4.set_title('d', loc='left', fontweight='bold', fontsize=fs_i)
ax4.set_title('DAV: %d'%dav_value_r30_1, loc='right', fontweight='bold', fontsize=fs)

ax5 = fig.add_axes([0.4, 0.45, 0.22, 0.15])
ax5.hist(dev_angle_r30_2.flatten(), bins=np.arange(-90, 92.5, 2.5), 
         edgecolor='black', color='black', align='mid', rwidth=1.0)
ax5.set_xlim(-90,90)
ax5.set_xlabel('deviation angle ($\\degree$)')
ax5.set_title('e', loc='left', fontweight='bold', fontsize=fs_i)
ax5.set_title('DAV: %d'%dav_value_r30_2, loc='right', fontweight='bold', fontsize=fs)

ax6 = fig.add_axes([0.7, 0.45, 0.22, 0.15])
ax6.hist(dev_angle_r30_3.flatten(), bins=np.arange(-90, 92.5, 2.5), 
         edgecolor='black', color='black', align='mid', rwidth=1.0)
ax6.set_xlim(-90,90)
ax6.set_xlabel('deviation angle ($\\degree$)')
ax6.set_title('f', loc='left', fontweight='bold', fontsize=fs_i)
ax6.set_title('DAV: %d'%dav_value_r30_3, loc='right', fontweight='bold', fontsize=fs)

ax7 = fig.add_axes([0.1, 0.08, 0.82, 0.25])
ax7.plot(data_tc['Date'], data_tc['Wind'], color='black', linewidth=3, label='TC Intensity')
ax7.set_ylim(20, 140)
ax7.set_xlabel('Date (UTC)')
ax7.set_ylabel('Wind (knots)', color='black')
ax7.set_title('g', loc='left', fontweight='bold', fontsize=fs_i)
ax8 = ax7.twinx()
ax8.plot(data_tc['Date'], data_tc['DAV 150 KM'], color='red', linewidth=3, label='DAV Radius 150 KM')
ax8.set_ylim(300, 3300)
ax8.set_ylabel('24-hour backward averaged\nDAV Value (degree$^2$)', color='black')
ax8.xaxis.set_major_locator(mdates.DayLocator(interval=2))

handles, labels = ax7.get_legend_handles_labels()  
handles2, labels2 = ax8.get_legend_handles_labels() 
handles = handles + handles2
labels  = labels + labels2
ax7.legend(handles, labels, loc='lower right', frameon=False, ncol=2)

if draft==False:
    plt.savefig("bolaven_case.png", dpi=200, bbox_inches='tight')
```
