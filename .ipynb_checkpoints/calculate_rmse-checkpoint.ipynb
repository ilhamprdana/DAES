# Import Modules

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.optimize import least_squares
plt.rcParams['font.sans-serif'] = 'Helvetica'
```

# Define Functions

```python
def sigmoid_func(dav, a, b, lc, uc):
    return lc + uc / (1 + np.exp(a * (dav - b)))

def error_func(par_fit, wind_data, dav, lc, u):
    a, b = par_fit
    wind_pred = sigmoid_func(dav, a, b, lc, uc)
    return wind_data/wind_pred-1

def rmse_func(wind_data, dav, a, b, lc, uc):
    wind_pred = sigmoid_func(dav, a, b, lc, uc)
    return np.sqrt(np.mean((wind_pred - wind_data) ** 2))
```

# Read Data

The data used in this study is the calculated DAV value, because including all netcdf data to be stored in the data availability might not be good option.

```python
data_tc_WP = pd.read_excel('./data/DAV_Smoothed_Western Pacific.xlsx')
```

# Calculation

This process will calculate the RMSE of the TC intensity estimates using DAV PMW. The cross-validation is used in this calculation, which means that a specified year will be used as test data while the remaining years data will be used as training data. In addition, the `scipy.optimize.least_square` is used to fit a sigmoid function to find the relationship between TC intensity as a function in DAV. The sigmoid function is: 

$f(\text{DAV}) = lc + \frac{uc}{1 + \exp(a \cdot (DAV - b))}$

where $a$ and $b$ are constants determined by the fitting, meanwhile $lc$ and $uc$ will set the lower bound and upper bound of the sigmoid. In this case we use $lc = 25$ and $uc = 140$ based on Ritchie et al. (2014) 

To calculate RMSE for a specified year of evaluation, just type the year in `year_eval` as well as the DAV radius in `header_dav_eval` variable. Repeat this process for the year and radius of calculation that you want.

```python
year_eval = 2023
a_init = 1e-3
b_init = 1000
lc = 25
uc = 140
header_dav_eval = "DAV75"
header_wind_eval = "TC Intensity"
header_date_eval = "Date"

test_data = data_tc_WP[data_tc_WP['Year'] == year_eval]
train_data = data_tc_WP[data_tc_WP['Year'] != year_eval]

fit_subset = least_squares(error_func, [a_init, b_init], args=[train_data[header_wind_eval], train_data[header_dav_eval], lc, uc])
a_fit, b_fit = fit_subset.x

rmse_val = rmse_func(test_data[header_wind_eval], test_data[header_dav_eval], a_fit, b_fit, lc, uc)
print("RMSE for test data %d on %s km is %.2f knots"%(year_eval, header_dav_eval, rmse_val))
```

# Plot (additional)

This section provides the process to vosualize the distribution of training data, test data, and the fitted sigmoid line. However, in order to produce better visualization, the test data is resampled every 6 hours (on synoptic time).

However, this plot is not shown in the extended abstract due to the page limitations.

```python
draft = True #True for drafting, and False for printing

fs = 14
plt.rcParams.update({'font.size': fs}) 
xmin, xmax = 500, 3500
ymin, ymax = 30, 160
cmin, cmax = 0, 5e-5

year_eval = 2023
a_init = 1e-3
b_init = 1000
lc = 25
uc = 140
header_dav_eval = "DAV75"
header_wind_eval = "TC Intensity"
header_date_eval = "Date"

test_data = data_tc_WP[data_tc_WP['Year'] == year_eval]
train_data = data_tc_WP[data_tc_WP['Year'] != year_eval]

dav_scheme = np.arange(0,4000)
test_data_6hourly = test_data[(test_data['Date'].dt.hour % 6 == 0) & (test_data['Date'].dt.minute == 0)]

fig = plt.figure(figsize=(6, 4))

hist = plt.hist2d(train_data[header_dav_eval], train_data[header_wind_eval], bins=[30,32], 
           range=[[xmin, xmax],[ymin,ymax]], cmap='ocean_r', density=True, vmin=cmin, vmax=cmax)
plt.plot(dav_scheme, sigmoid_func(dav_scheme, a_fit, b_fit, lc, uc), linewidth=3, color='black', label='Fitted Sigmoid')
plt.scatter(test_data_6hourly[header_dav_eval], test_data_6hourly[header_wind_eval],
           marker='x', s=20, color='orange', label = "Test Data (%s)"%(str(year_eval)))
plt.title('RMSE : %.2f knots'%rmse_func(test_data[header_wind_eval], test_data[header_dav_eval], a_fit, b_fit, lc, uc), loc='left', fontsize=fs)
plt.xlabel("DAV Value (degree$^2$)", fontsize=fs)
plt.ylabel("TC Intensity (knots)", fontsize=fs)
plt.legend(frameon=False, fontsize=fs)

cbar = plt.colorbar(hist[3], extend='max')  # hist[3] is the QuadMesh object needed for the colorbar
cbar.set_label("Density")  # Set label for the colorbar

if draft==False:
    plt.savefig("test_training_sigmoid_visualization.png", dpi=200, bbox_inches='tight')
```
